##Converting the categorical features into numerical.

```python
from sklearn.preprocessing import OneHotEncoder
cat_features = df2.select_dtypes(include='object')
num_features = df2.select_dtypes(exclude='object')
df2 = pd.get_dummies(df2, drop_first=True)
df2
```

# Training Model

```python
from sklearn.model_selection import train_test_split
X = df2.drop(['ASD_TRAITS_YES'], axis=1)
y = df2['ASD_TRAITS_YES']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=4)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X
```

```python
y
```

```python
##Logistics regression
```

```python
#logistic method
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=1000, C=1.0)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
logreg.score(X_train, y_train)
```

```python
# calculate R2 score
from sklearn.metrics import r2_score
logreg_r2 = r2_score(y_test, y_pred)
print('R2 score:',logreg_r2)
```

```python
#classification report
#confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
c_report = classification_report(y_test, y_pred)
print('Classification Report\n', c_report )
```

```python
confusion_matrix(y_test,y_pred)
```

# K-FOLD

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
logreg = LogisticRegression(max_iter=1000, C=1.0)
kfold_validation=KFold(10)
```

```python
from sklearn.model_selection import cross_val_score
results=cross_val_score(logreg,X,y,cv=kfold_validation)
print(results)
print(np.mean(results))
```

# Stratified K-fold Cross Validation

```python
from sklearn.model_selection import StratifiedKFold
skfold=StratifiedKFold(n_splits=5)
logreg = LogisticRegression(max_iter=1000, C=1.0)
scores=cross_val_score(logreg,X,y,cv=skfold)
print(np.mean(scores))
```
