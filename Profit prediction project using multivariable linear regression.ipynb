#Separating independent(x) & Dependent(y) value

```python
# Here, the independent variables are Marketing spend, Administration, Transportation, Area.
# and, the dependent variable is profit.
#this line will assign all the values to 'X' except profit column bcz profit column will be droped.
```

```python
x = df.drop(["Profit"],axis=1)
```

```python
x
```

```python
y = df["Profit"]
```

```python
y
```

ONE HOT ENCODING

```python
# Here, we will have to do 'one hot encoding', not 'lebel encoding', bcz to do lebel encoding the data must have to be "ordered data"
```

```python
#convert the column into categorical column
```

```python
city = pd.get_dummies(x["Area"],drop_first=True)
```

```python
city
```

```python
#drop the Area column to remove the categorical value and add the numerical value columns we got by one hot encoder.
x = x.drop('Area',axis=1)
```

```python
x
```

```python
#concatanation
x = pd.concat([x,city],axis=1)
```

```python
x
```

```python
#splitting the data set into test set & training set.
```

```python
#import libraby
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#Fitting multiple linear regression to the training set
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(xtrain, ytrain)
```

```python
ytest
```

```python
#predicting the test set
ypred = regressor.predict(xtest)
```

```python
ypred
```

R-SQUARED VALUE

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(ytest,ypred)
```

```python
score
```

```python
#we can also have the accuracy another way
regressor.score(xtest, ytest)
```
